# Hypothesis: 
1. AutoDrive Components Ltd. – Engine Assembly Time Evaluation

```python
# H0
# H1
```

```python
theory_mean = 1.8 + 0.02*450
mean = 10.8
std = 25
sample = 35
alpha = 0.05
```

```python
# tstats
```

```python
import numpy as np
from scipy.stats import norm
```

```python
standard_error = std/np.sqrt(sample)
t_statistics = (mean - theory_mean)/standard_error
t_statistics
```

```python
# critical_value
```

```python
critical_value = norm.ppf(1-alpha)
critical_value
```

```python
# Make a Decision
```

```python
if t_statistics > critical_value:
   decision = "H1"
else:
   decision = "H0"
```

```python
decision
```
